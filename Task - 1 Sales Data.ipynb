##  Standardize text values 

```python
# Clean 'Customer_Gender' and 'Country' columns
data['Customer_Gender'] = data['Customer_Gender'].str.strip().str.lower()
data['Country'] = data['Country'].str.strip().str.title()
```

- **.str.strip()** --> removes leading/trailing spaces
- **.str.lower()** → makes all text lowercase (e.g., "M", "F" → "m", "f")
- **.str.upper()** → makes all text uppercase
- **.str.title()** → capitalizes each word (e.g., "united states" → "United States")

```python
data
```

## Convert date formats 

```python
# Step 1: Ensure the column is in datetime format
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Step 2: Convert to dd-mm-yyyy format (as string)
data['Date'] = data['Date'].dt.strftime('%d-%m-%Y')
```

```python
data
```

## Renaming column headers

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
data 
```

- **.str.strip()** → removes leading/trailing whitespace
- **.str.lower()** → converts to lowercase
- **.str.replace(' ', '_')** → replaces spaces with underscores
